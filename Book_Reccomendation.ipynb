```python
def rec(name):
    # Find the index of the given name in the pivot table
    id = np.where(pivot.index == name)[0][0]
    
    # Get the distances and indices of the nearest neighbors
    distances, suggestions = model.kneighbors(pivot.iloc[id, :].values.reshape(1, -1), n_neighbors=6)
    
    # Loop through the suggestions and print the index, skipping the first one (itself)
    for i in range(1, len(suggestions[0])):  # Start at 1 to skip the original item
        print(pivot.index[suggestions[0][i]])

# Call the function with an example
rec("Animal Farm")
```
